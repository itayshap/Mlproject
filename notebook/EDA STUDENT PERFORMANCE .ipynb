### 3.7 Exploring Data

```python
print("Categories in 'gender' variable:  ", df['gender'].unique())

print("Categories in 'race_ethnicity' variable:  ",df['race_ethnicity'].unique())

print("Categories in'parental level of education' variable:",df['parental_level_of_education'].unique())

print("Categories in 'lunch' variable: ",df['lunch'].unique())

print("Categories in 'test preparation course' variable: ",df['test_preparation_course'].unique())
```

```python
numerical_df = df.select_dtypes(include=['number'])
numeric_features = numerical_df.columns.to_list()
categorical_df = df.select_dtypes(exclude=['number'])
categorical_features = categorical_df.columns.to_list()

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df.loc[:,numeric_features].sum(axis=1)
df['average'] = df.loc[:,numeric_features].mean(axis=1)
df.head()
```

```python
math_full = df.loc[df.math_score == 100, "average"].count()
writing_full = df.loc[df.writing_score == 100, "average"].count()
reading_full = df.loc[df.reading_score == 100, "average"].count()
print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 =  df.loc[df.reading_score <= 20 , "average"].count()
writing_less_20 = df.loc[df.writing_score <= 20 , "average"].count()
math_less_20 = df.loc[df.math_score <= 20 , "average"].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#### 4.4.4 LUNCH COLUMN 
- Which type of lunch is most common amoung students ?

#### UNIVARIATE ANALYSIS ( Which type of lunch is most common amoung students ? )

```python
plt.rcParams['figure.figsize'] = (8, 4)
sns.countplot(df['lunch'], palette = 'pastel')
plt.title('Comparison of different types of lunch', fontsize = 15)
plt.xlabel('types of lunch')
plt.ylabel('count')
plt.show()
```

#### Insights 
- Students being served Standard lunch was more than free lunch

#### BIVARIATE ANALYSIS (  Is lunch type intake has any impact on student's performance ? )

```python

average_scores = df[['math_score', 'reading_score', 'writing_score', 'lunch']].groupby('lunch').mean().reset_index()
# Melting the DataFrame for plotting
df_melted = average_scores.melt(id_vars="lunch", var_name="score_type", value_name="average score")
# Plot using seaborn
plt.figure(figsize=(8, 5))
g= sns.barplot(data=df_melted, x="score_type", y="average score", hue="lunch")
for container in g.containers:
    g.bar_label(container,color='black')
```

#### Insights 
- Students who get Standard Lunch tend to perform better than students who got free/reduced lunch

#### 4.4.5 TEST PREPARATION COURSE COLUMN 
- Is Test prepration course has any impact on student's performance ?

```python

average_scores = df[['math_score', 'reading_score', 'writing_score', 'test_preparation_course']].groupby('test_preparation_course').mean().reset_index()
# Melting the DataFrame for plotting
df_melted = average_scores.melt(id_vars="test_preparation_course", var_name="score_type", value_name="average score")
# Plot using seaborn
plt.figure(figsize=(8, 5))
g= sns.barplot(data=df_melted, x="score_type", y="average score", hue="test_preparation_course")
for container in g.containers:
    g.bar_label(container,color='black')
```

#### Insights  
- Students who have completed the Test Prepration Course have scores higher in all three categories than those who haven't taken the course
